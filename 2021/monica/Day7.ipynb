```python
file = 'example_day7.txt'
with open (file) as f:
    data = f.read().split('\n')
    data = [ d.split(',') for d in data if d != ''][0]
    data = [ int(d) for d in data]
data
```

```python
costs = []
for pos in range(0,len(data)+1):
    data2 = [ abs(d-pos) for d in data]
    costs.append(sum(data2))
print(min(costs))
costs
```
