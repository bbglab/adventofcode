```python
import operator
import numpy as np
from itertools import product
from functools import reduce

# with 3 dimensions

with open('input.txt', 'rt') as f:
    grid = np.array(list(map(list, f.read().splitlines())))
n, m = grid.shape 
active3d = [(x,y,0) for x,y in product(range(n), range(m)) if grid[x,y]=='#']
```

```python
def adjacent_count(coords, *p):
    a = sum([
        reduce(operator.and_, 
               list(map(lambda t: abs(t[0]-t[1])<=1, 
                        tuple(zip(p, c))))) for c in coords
            ])
    return a - (p in coords)

def spawn(coords):
    n = len(coords[0])
    acc, visited = [], []
    for s in coords:
        for t in product([1,0,-1], repeat=n):
            p = tuple(map(sum, tuple(zip(s,t))))
            if p not in visited:
                count = adjacent_count(coords, *p)
                if ((p in coords) and (2 <= count <= 3)) or ((p not in coords) and (count == 3)):
                    acc.append(p)
                visited.append(p)
    return acc

def cycles(coords, n):
    for _ in range(n):
        coords = spawn(coords)
    return len(coords)
```

```python
# test

test_state = [(1,0,0),(2,1,0),(0,2,0),(1,2,0),(2,2,0)]
cycles(test_state, 6)
```
